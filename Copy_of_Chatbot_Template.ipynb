#Speaking
Below is a function that will take a response and turn it into a wav file.  After creating the wav file it will be able to play the sound it creates.

```
from gtts import gTTS
import IPython.display 
sound_file2 = '2.wav'
def speak(response):
  question = gTTS(response)
  question.save('2.wav')
  sound_file2 = '2.wav'
  IPython.display.display(IPython.display.Audio(sound_file2, autoplay=True, embed=True)) 
  print (response)
```

# Thinking
Our chat algorithm needs to come up with responses to words.  This is how it gets done.  This set of rules will respond to both input from the user and from the circuit playground express. Some basic rules and responses are here.  You can add your own.

```
def think(question):
  if ("food" in question or "breakfast" in question or "lunch" in question or "dinner" in question):
    speak('I am a robot.  I do not eat anything, but I am glad you can enjoy food, human.')
  if ("name" in question):
    speak("My name is Chatbot3000")
  if ("think" in question):
    speak("Well, I don't so much think as compute")
  if (question == "Bright Light"):
    speak("Good Morning, Time to get to work!")
while True:
  user_question = input ("What is your question?")
  think(user_question)
```
