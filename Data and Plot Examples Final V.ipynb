# **Plot Style 3**

```
style.use('dark_background') #this turns the background area of the chart to dark

plt.rcParams["figure.figsize"] = [15, 10]#Scales the size of the chart
fig, ax1 = plt.subplots()# Generates the chart area

ax1.plot(figsize=(10,5))
ax1.title.set_text('Acceleration vs Time')#Chart Title
ax1.plot(x, y1, 'g-',)# plots the data points of a selected data set
ax1.set_xlabel("Time (s)")#Labels x-axis
ax1.set_ylabel("Absolute acceleration (m/s^2)")#Labels y-axis

# Don't allow the axis to be on top of your data
ax1.set_axisbelow(True)

# Turn on the minor TICKS, which are required for the minor GRID
ax1.minorticks_on()

# Customize the major grid
ax1.grid(which='major', linestyle='-', linewidth='0.5', color='red')
# Customize the minor grid
ax1.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Turn off the display of all ticks.
ax1.tick_params(which='both', # Options for both major and minor ticks
                top='off', # turn off top ticks
                left='off', # turn off left ticks
                right='off',  # turn off right ticks
                bottom='off') # turn off bottom ticks
```
